```python
import numpy as np
import pandas as pd

def normalize_l2(x):
    x = np.array(x)
    if x.ndim == 1:
        norm = np.linalg.norm(x)
        if norm == 0:
            return x
        return x / norm
    else:
        norm = np.linalg.norm(x, 2, axis=1, keepdims=True)
        return np.where(norm == 0, x, x / norm)
```

```python
ss0 = ['anatomy','clinical_knowledge','college_biology', 'college_medicine','medical_genetics']

for ss in ss0:
    df = pd.read_csv(f"data/questions_{ss}.csv")
    df["embedding"] = df["embedding"].apply(lambda x: normalize_l2(np.array(eval(x))[:256]).tolist())
    df.to_csv(f"data/reduced_questions_{ss}.csv", index=False)
```
